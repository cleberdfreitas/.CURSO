### HTML

```python
import pandas as pd
url = 'https://raw.githubusercontent.com/alura-cursos/Pandas/refs/heads/main/filmes_wikipedia.html'
dados_html = pd.read_html(url)
dados_html
```

```python
type(dados_html)
```

```python
len(dados_html)
```

```python
top_filmes = dados_html[1]
top_filmes
```

```python
top_filmes.to_html('top_filmes.html')
pd.read_html('top_filmes.html')
```

```python
top_filmes.to_csv('top_filmes_1998.csv', index=False)
pd.read_csv('top_filmes_1998.csv')
```
